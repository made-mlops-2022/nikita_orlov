Построим корреляционную матрицу, чтобы узнать линейную корреляцию признаков:

```python
import seaborn as sns;sns.set(style='whitegrid')
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = (10,8)

def plot_corr_matr(df_data):
    corr_raw = df_data.corr().round(2)
    mask = np.triu(np.ones_like(corr_raw))
    corr = corr_raw.iloc[1:, :-1]
    mask = mask[1:, :-1]
    sns.heatmap(corr, mask=mask, square=False, cbar=False, annot=True)
```

```python
plot_corr_matr(df_raw)
```

Целевая переменная больше всего зависит от **thalach** (обратная зависимость), **exang**, **oldpeak**, **ca**, **thal**

```python
def plot1count_ordered(df, x, xlabel):
    plt.figure(figsize=(20,2))
    sns.countplot(x=x,hue='condition',data=df)
    plt.legend(["0", "1"],loc='upper right')
    plt.xlabel(xlabel)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plot1count_ordered(df_raw, "age", "Age")
```

До 55 лет реже болеют и с 71-77 лет не болеют

```python
plot1count_ordered(df_raw, "oldpeak", "oldpeak")
```

```python
''' Draw a Bivariate Seaborn Pairgrid /w KDE density w/ '''
def snsPairGrid(df):

    ''' Plots a Seaborn Pairgrid w/ KDE & scatter plot of df features'''
    g = sns.PairGrid(df,diag_sharey=False,hue='condition',palette='Purples')
    g.fig.set_size_inches(13,13)
    g.map_upper(sns.kdeplot,n_levels=5)
    g.map_diag(sns.kdeplot, lw=2)
    g.map_lower(sns.scatterplot,s=20,edgecolor="k",linewidth=1,alpha=0.6)
    g.add_legend()
    plt.tight_layout()
```

```python
numvars_targ = ['age','trestbps','chol','thalach','oldpeak','condition']
snsPairGrid(df_raw[numvars_targ])
```

Распределение признаков **chol** и **trestbps** выглядит практически идентично (что подтверждает корреляционная матрица)

Распределение признаков **oldpeak** и **thalah** различается сильнее всего
